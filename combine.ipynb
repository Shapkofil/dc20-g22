```python
filepath_wa = "./data/Metropolitan-street.parquet"

# read the Parquet file into a pandas dataframe
df_ward = pq.read_pandas(filepath_wa).to_pandas()
barnet_df = df_ward[df_ward['LSOA name'].str.contains('Barnet')==True]
barnetBur_df = barnet_df[barnet_df['Crime type'].str.contains('Burglary')==True]

barnet_df
```

```python
df_occupation = pd.read_csv("./data/Economic_Activity.csv")

df_occupation.iloc[:,4:] = df_occupation.iloc[:,4:].div(df_occupation['All usual residents aged 16-74 '], axis=0)
df_occupation = df_occupation.drop(['local authority code', 'local authority name', 'All usual residents aged 16-74 '], axis=1)


barnetBur_df = pd.merge(left=barnetBur_df, right=df_occupation, left_on='LSOA code', right_on='LSOA code')
#barnetBur_df
```

```python
df_ethnic_group = pd.read_csv("./data/Ethnic_group.csv")


df_ethnic_group.iloc[:,4:] = df_ethnic_group.iloc[:,4:].div(df_ethnic_group['All usual residents'], axis=0)

df_ethnic_group = df_ethnic_group.drop(['local authority code', 'local authority name', 'All usual residents', 'Unnamed: 7'], axis=1)
df_ethnic_group

barnetBur_df = pd.merge(left=barnetBur_df, right=df_ethnic_group, left_on='LSOA code', right_on='LSOA code')
#barnetBur_df
```

```python






encoded_wards = pd.get_dummies(df_final['WD22NM'])
df_months = pd.get_dummies(df_final['Month'].str.split("-").str[-1].astype(int))
df_months.columns = ['jan', 'feb', 'mar', 'apr', 'may', 'jun', 'jul', 'aug', 'sep', 'okt', 'nov', 'dec']
encoded_wards.columns
df_mon = df_months.join(encoded_wards)
df_mon
```

```python

df_crimes = df_final[["Month", "Crime ID","WD22NM"]].copy()
df_crimes= df_crimes.groupby(["WD22NM","Month"]).count()


def rolling_sum2(group,y,str,s):
    group[str] = group['Crime ID'].rolling(window=y).sum().shift(s)
    return group

df_crimes = df_crimes.groupby('WD22NM', as_index=False).apply(lambda x: rolling_sum2(x,1,"1Months",1))
df_crimes.fillna(0, inplace=True)
df_crimes = df_crimes.groupby('WD22NM', as_index=False).apply(lambda x: rolling_sum2(x,2,"2Months",1))
df_crimes.fillna(0, inplace=True)
df_crimes = df_crimes.groupby('WD22NM', as_index=False).apply(lambda x: rolling_sum2(x,3,"3Months",1))
df_crimes.fillna(0, inplace=True)
df_crimes = df_crimes.groupby('WD22NM', as_index=False).apply(lambda x: rolling_sum2(x,4,"4Months",1))
df_crimes.fillna(0, inplace=True)
df_crimes = df_crimes.groupby('WD22NM', as_index=False).apply(lambda x: rolling_sum2(x,5,"5Months",1))
df_crimes.fillna(0, inplace=True)
df_crimes = df_crimes.groupby('WD22NM', as_index=False).apply(lambda x: rolling_sum2(x,6,"6Months",1))
df_crimes.fillna(0, inplace=True)
df_crimes = df_crimes.groupby('WD22NM', as_index=False).apply(lambda x: rolling_sum2(x,7,"7Months",1))
df_crimes.fillna(0, inplace=True)
df_crimes = df_crimes.groupby('WD22NM', as_index=False).apply(lambda x: rolling_sum2(x,8,"8Months",1))
df_crimes.fillna(0, inplace=True)
df_crimes = df_crimes.groupby('WD22NM', as_index=False).apply(lambda x: rolling_sum2(x,9,"9Months",1))
df_crimes.fillna(0, inplace=True)
df_crimes = df_crimes.groupby('WD22NM', as_index=False).apply(lambda x: rolling_sum2(x,10,"10Months",1))
df_crimes.fillna(0, inplace=True)
df_crimes = df_crimes.groupby('WD22NM', as_index=False).apply(lambda x: rolling_sum2(x,11,"11Months",1))
df_crimes.fillna(0, inplace=True)
df_crimes = df_crimes.groupby('WD22NM', as_index=False).apply(lambda x: rolling_sum2(x,12,"12Months",1))
df_crimes.fillna(0, inplace=True)




#Merge the count of crimes to the main dataframe
merged_df = pd.merge(df_final, df_crimes, on=['Month', 'WD22NM'], how='left')



merged_df["Month"]=pd.to_datetime(merged_df["Month"])
merged_df["Year"]=merged_df["Month"].dt.year

pd.set_option('display.max_columns', None)
merged_df=merged_df.join(df_mon)

merged_df['Unnamed: 2'] = merged_df['Unnamed: 2'].str.replace('%', '')
merged_df['Unnamed: 2'] = merged_df['Unnamed: 2'].str.replace(',', '.').astype(float)


merged_df = merged_df.drop(['Crime ID_x', 'Month', 'Reported by', 'Falls within', 'Longitude', 'Latitude', 'LSOA code', 'Location', 'LSOA name',
                            'Last outcome category', 'Context', 'Crime type', 'Unnamed: 1', 'WD22NM'], axis=1)
merged_df.head()
```
