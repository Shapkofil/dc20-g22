```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import pyarrow as pa
import pyarrow.parquet as pq

from sklearn.preprocessing import StandardScaler

# Specify the path to your Parquet file



file_path = "./data/linearmodel.parquet"

# Read the Parquet file into a pandas DataFrame
df = pd.read_parquet(file_path)
onehot_df = df.iloc[:, [1] + list(range(64-23,64))]
month_embeds = onehot_df.groupby("Month").sum()
# month_totals = month_embeds.sum(axis=1)
# month_pdf = month_embeds.apply(lambda row: row / month_totals[row.name], axis=1) * 2.0 -1
# #month_pdf
# scaler = StandardScaler()
# normalized_df = scaler.fit_transform(month_embeds)
# normalized_df = pd.DataFrame(normalized_df, columns=month_embeds.columns)
# normalized_df
#normalized_df.to_parquet( "./data/lstm2.parquet")
#month_pdf
month_embeds[-12:]
```

Min-Max Scaling


```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
minmax_df = scaler.fit_transform(month_embeds)
minmax_df = pd.DataFrame(minmax_df, columns=month_embeds.columns)
minmax_df
#minmax_df.to_parquet( "./data/lstm_minmax.parquet")
```

Max Abs Scaler

```python
from sklearn.preprocessing import MaxAbsScaler
scaler = MaxAbsScaler()
maxabs_df = scaler.fit_transform(month_embeds)
maxabs_df = pd.DataFrame(maxabs_df, columns=month_embeds.columns)
maxabs_df.to_parquet( "./data/lstm_maxabs.parquet")
```

RobustScaler


```python
from sklearn.preprocessing import RobustScaler
scaler = RobustScaler()
robust_df = scaler.fit_transform(month_embeds)
robust_df = pd.DataFrame(robust_df, columns=month_embeds.columns)
robust_df.to_parquet( "./data/lstm_robust.parquet")
```

Quantile Transformer Scaler / Rank Scaler


```python
from sklearn.preprocessing import QuantileTransformer
scaler = QuantileTransformer()
rank_df = scaler.fit_transform(month_embeds)
rank_df = pd.DataFrame(rank_df, columns=month_embeds.columns)
#rank_df.to_parquet( "./data/lstm_rank.parquet")
```

Power Transformer ('yeo-johnson')

```python
from sklearn.preprocessing import PowerTransformer
scaler = PowerTransformer(method='yeo-johnson')
power_df = scaler.fit_transform(month_embeds)
power_df=pd.DataFrame(power_df,columns=month_embeds.columns)
#power_df.to_parquet( "./data/lstm_power.parquet")
power_df
```

Unit Vector Scaler

```python
magnitudes = np.linalg.norm(month_embeds, axis=1)  # Calculate magnitudes of each vector
unit_df = month_embeds / magnitudes[:, np.newaxis]
unit_df.reset_index(inplace=True)
unit_df.drop("Month", axis=1, inplace=True)
unit_df.to_parquet( "./data/lstm_unit.parquet")
```

```python
from sklearn import preprocessing
df_norm  = preprocessing.normalize(month_embeds, norm='l2')
df_norm=pd.DataFrame(df_norm,columns=month_embeds.columns)
df_norm.to_parquet( "./data/lstm_norm.parquet")
```
